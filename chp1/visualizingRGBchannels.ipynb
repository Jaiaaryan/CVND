# RGB colorspace
### Import resource

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline 
# this makes all the matplot lib to embed inside the code
```

### Read in an image

```python
image = mpimg.imread('car.jpeg')
plt.imshow(image)
```

### RGB channels
Visualize the level of each **_color_** channel
Opencv Reads the image as BGR where as Matplotlib reads images as RGB

```python
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

# visualize the individual color channel
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20,10))
ax1.imshow(r,cmap = 'gray')
ax2.imshow(g,cmap = 'gray')
ax3.imshow(b,cmap = 'gray')
```

Find which area has the lowest value for the channel **_red_** and **_blue_**

```python
red_min = np.amin(r)
blue_min = np.amin(b)
print('r_min:',red_min,'b_min', blue_min)
```

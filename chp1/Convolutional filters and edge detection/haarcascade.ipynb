# Face detection using Haar cascades


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread(r'D:\00_Python Scripts\Test\udacity_nano_computervision\CVND_Exercises-master\1_2_Convolutional_Filters_Edge_Detection\images\multi_faces.jpg')

# convert to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.figure(figsize=(20,10))
plt.imshow(gray, cmap = 'gray')

```

 Next we load a fully trained architecture of the face detector in the file detector_architectures/haarcascade_frontalface_default.xml and use it to find faces
 
**A note on parameters**
How many faces are detected by the function?
- detectMultiScale aims to detect faces of varying sizes.
- Input of this function are
 - image 
 - scaleFactor
 - minNeighbors
 

```python
# load the cascade classifier:
face_cascade = cv2.CascadeClassifier(r"C:\Users\jsrinivasan\Downloads\opencv\build\etc\haarcascades\haarcascade_frontalface_default.xml")
# run this detector on the gray scale image
faces = face_cascade.detectMultiScale(gray,4,6)
```

```python
print('We found'+ str(len(faces))+'faces in this image')
print('Their coordinates and length/width are as follows')
print('============================================================================')
print(faces)
```

```python
img_with_detections = image.copy()

for x,y,w,h in faces:
    cv2.rectangle(img_with_detections,(x,y),(x+w,y+h),(255,0,0),5)
    
plt.figure(figsize=(20,10))
plt.imshow(img_with_detections)
```

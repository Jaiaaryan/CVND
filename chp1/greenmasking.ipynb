# Masking Green color
### import packages


```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

### Read in the image and change color

```python
image = cv2.imread('car_green_screen.jpg')
# print image shape and size 
print(image.shape)
# Make a copy of the image
image_copy= np.copy(image)
# display the image 
plt.imshow(image_copy)
```

### create a mask
display only the car and the background as black


```python
lower_green = np.array([10,150,10])
upper_green = np.array([80,255,90])
```

 the boundaries are defined, use inRange function to make all the pixels white if they are lying inside the range,
 all other pixels that doesn't lie within this range make it black
 

```python
mask = cv2.inRange(image_copy, lower_green, upper_green)
# display the mask
plt.imshow(mask, cmap= 'gray')
```

now show an image that has only the car in it, without the background


```python
only_car = image.copy()
only_car[mask != 0 ]= [0,0,0]
# display the masked_image
plt.imshow(only_car)
 
```

### include a background image and put the car in it


```python
bg_image = cv2.imread('water_balloons.jpg')
bg_image= bg_image[0:450,0:660]
bg_image = cv2.cvtColor(bg_image, cv2.COLOR_BGR2RGB)
# display bg 
plt.imshow(bg_image)
```

```python
bg_image[mask == 0] = [0,0,0]
# display the bg image cut out for the car profile
plt.imshow(bg_image)
```

### Add car to the background image


```python
new_image = bg_image + only_car
plt.imshow(new_image)
```
